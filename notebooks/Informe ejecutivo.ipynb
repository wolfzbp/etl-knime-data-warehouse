## **🎯 1. Objetivo**

El objetivo del presente proyecto fue diseñar e implementar un flujo **ETL (Extract, Transform, Load)** utilizando **KNIME**, integrando múltiples fuentes de datos heterogéneas y consolidando la información en un **Data Warehouse relacional**.

Este almacén fue estructurado mediante un **esquema estrella**, permitiendo realizar análisis **OLAP (Online Analytical Processing)** sobre información de ventas.



---

## **🗂️ 2. Fuentes de Datos**

Se utilizaron **cuatro fuentes de datos** de distinta naturaleza, simulando un entorno real de integración:

- 📄 **CSV:** archivo `ventas.csv` con transacciones de ventas  
- 🐬 **MySQL:** tabla `clientes` con información de clientes  
- 🐘 **PostgreSQL:** tabla `productos` con catálogo de productos  
- 🍃 **MongoDB:** colección `tiendas` con datos semiestructurados  

> 💡 Los datos fueron generados de manera sintética mediante **Python y la librería Faker**, y cargados a las bases de datos previamente mencionadas, permitiendo trabajar con estructuras realistas. La ruta hacia el .ipynb se muestra a continuacion:
[generacion_datos.png](..\notebooks\generacion_datos.ipynb)



---

## **🔄 3. Proceso ETL**

El flujo ETL fue desarrollado en **KNIME** y se estructuró en tres etapas:

### **📥 3.1 Extracción**

Se establecieron conexiones a múltiples fuentes mediante nodos especializados:

- `CSV Reader`
- `DB Query Reader`
- `MongoDB Connector`
- `JSON to Table`

---

### **🔧 3.2 Transformación**

Se integraron las distintas fuentes mediante **joins**, generando un dataset consolidado.

Además, se realizaron transformaciones clave:

- 🔗 Integración de datos entre múltiples fuentes  
- 🧮 Cálculo de métricas de negocio  
- 📅 Transformación de fechas  
- 🔄 Agrupación de datos  
- 🧹 Eliminación de duplicados  

---

### **📤 3.3 Carga**

Los datos transformados fueron cargados en **PostgreSQL (`dw_retail`)** mediante nodos:

- `DB Writer`

Esto permitió construir un **Data Warehouse funcional**.

[Ruta al KNIME Workflow](..\workflows\ETL_Retail_DW.knwf)

![Worflow completo](../diagrams/KNIME_worflow.png)

## **🧹 4. Transformación y Limpieza de Datos**

Durante esta etapa se aplicaron técnicas para mejorar la calidad de los datos:

### **✔ Procesos realizados**

- 🔍 Filtrado de registros (`Row Filter`)
- 🚫 Eliminación de duplicados (`Duplicate Row Filter`)
- 🔄 Conversión de tipos (`String to Date&Time`)
- 📆 Extracción de componentes temporales (`Date&Time Extractor`)
- 📊 Agrupación de datos (`GroupBy`)

---

### **📈 Métricas calculadas**

Se generaron indicadores clave de negocio:

- **Venta Bruta**
- **Venta Neta**
- **Costo Total**
- **Utilidad**

> Estas transformaciones permiten obtener información lista para análisis.

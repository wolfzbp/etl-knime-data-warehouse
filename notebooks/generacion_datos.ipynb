# Generacion de datos sinteticos

````python
import pandas as pd
import numpy as np
import random
import json
from faker import Faker
from datetime import datetime, timedelta
````

````python
fake = Faker("es_ES")
random.seed(42)
np.random.seed(42)

N_CLIENTES = 300
N_PRODUCTOS = 200
N_TIENDAS = 20
N_EMPLEADOS = 20
N_VENTAS = 2000
````

````python
PROVINCIAS_CIUDADES = {
    "Pichincha": ["Quito", "Cayambe", "Rumiñahui"],
    "Guayas": ["Guayaquil", "Durán", "Samborondón"],
    "Azuay": ["Cuenca", "Gualaceo"],
    "Manabí": ["Manta", "Portoviejo"],
    "Tungurahua": ["Ambato", "Baños"],
    "El Oro": ["Machala", "Santa Rosa"]
}

SEGMENTOS = ["Frecuente", "Nuevo", "Corporativo", "Premium"]
GENEROS = ["Masculino", "Femenino", "No especificado"]

CATEGORIAS = {
    "Electrónica": ["Audio", "Computación", "Telefonía"],
    "Hogar": ["Cocina", "Decoración", "Muebles"],
    "Deportes": ["Fitness", "Outdoor", "Accesorios"],
    "Moda": ["Ropa", "Calzado", "Accesorios"],
    "Oficina": ["Papelería", "Mobiliario", "Tecnología"]
}

MARCAS = [
    "Samsung", "LG", "Sony", "Xiaomi", "HP", "Lenovo",
    "Ikea", "Nike", "Adidas", "Mabe", "Whirlpool", "Genérica"
]

ESTADOS_PRODUCTO = ["activo", "inactivo", "descontinuado"]
TIPOS_TIENDA = ["Fisica", "Outlet", "Express", "Mall"]
SERVICIOS = ["Entrega a domicilio", "Retiro en tienda", "Soporte técnico", "Parqueadero", "Pago QR"]
METODOS_PAGO = ["Tarjeta", "Efectivo", "Transferencia", "PayPal"]
CANALES_VENTA = ["Online", "Presencial", "App"]
ESTADOS_VENTA = ["completada", "completada", "completada", "anulada"]  # sesgo intencional

CARGOS = ["Vendedor", "Supervisor", "Cajero", "Asesor comercial"]
````

````python
empleados = []
for i in range(1, N_EMPLEADOS + 1):
    provincia = random.choice(list(PROVINCIAS_CIUDADES.keys()))
    ciudad = random.choice(PROVINCIAS_CIUDADES[provincia])

    empleados.append({
        "id_empleado": i,
        "nombre_empleado": fake.name(),
        "cargo": random.choice(CARGOS),
        "canal_base": random.choice(CANALES_VENTA),
        "ciudad_base": ciudad,
        "provincia_base": provincia
    })

df_empleados = pd.DataFrame(empleados)
df_empleados.head()
````

````python
clientes = []
for i in range(1, N_CLIENTES + 1):
    provincia = random.choice(list(PROVINCIAS_CIUDADES.keys()))
    ciudad = random.choice(PROVINCIAS_CIUDADES[provincia])

    clientes.append({
        "id_cliente": i,
        "nombre": fake.first_name(),
        "apellido": fake.last_name(),
        "email": fake.unique.email(),
        "telefono": fake.phone_number(),
        "ciudad": ciudad,
        "provincia": provincia,
        "segmento": random.choice(SEGMENTOS),
        "fecha_registro": fake.date_between(start_date="-3y", end_date="today"),
        "edad": random.randint(18, 75),
        "genero": random.choice(GENEROS)
    })

df_clientes = pd.DataFrame(clientes)
df_clientes.head()
````

````python
productos = []
for i in range(1, N_PRODUCTOS + 1):
    categoria = random.choice(list(CATEGORIAS.keys()))
    subcategoria = random.choice(CATEGORIAS[categoria])
    marca = random.choice(MARCAS)

    precio_lista = round(random.uniform(8, 1500), 2)
    costo_unitario = round(precio_lista * random.uniform(0.45, 0.80), 2)

    productos.append({
        "id_producto": i,
        "sku": fake.unique.bothify(text="SKU-#####"),
        "nombre_producto": f"{fake.word().capitalize()} {fake.word().capitalize()}",
        "categoria": categoria,
        "subcategoria": subcategoria,
        "marca": marca,
        "precio_lista": precio_lista,
        "costo_unitario": costo_unitario,
        "stock": random.randint(0, 300),
        "estado_producto": random.choice(ESTADOS_PRODUCTO),
        "fecha_creacion": fake.date_between(start_date="-2y", end_date="today")
    })

df_productos = pd.DataFrame(productos)
df_productos.head()
````

````python
tiendas = []

for i in range(1, N_TIENDAS + 1):
    provincia = random.choice(list(PROVINCIAS_CIUDADES.keys()))
    ciudad = random.choice(PROVINCIAS_CIUDADES[provincia])

    if provincia in ["Guayas", "Manabí", "El Oro"]:
        region = "Costa"
    else:
        region = "Sierra"

    tienda = {
        "id_tienda": i,
        "nombre_tienda": f"Tienda {fake.company()}",
        "ciudad": ciudad,
        "provincia": provincia,
        "region": region,
        "tipo_tienda": random.choice(TIPOS_TIENDA),
        "horarios": {
            "lunes_viernes": "09:00-18:00",
            "sabado": "10:00-16:00",
            "domingo": random.choice(["cerrado", "10:00-14:00"])
        },
        "contacto": {
            "telefono": fake.phone_number(),
            "email": fake.company_email()
        },
        "servicios": random.sample(SERVICIOS, k=random.randint(2, 5)),
        "rating": round(random.uniform(3.2, 5.0), 1),
        "geo": {
            "lat": round(random.uniform(-4.5, 0.5), 6),
            "lon": round(random.uniform(-81.0, -75.0), 6)
        }
    }
    tiendas.append(tienda)

len(tiendas), tiendas[0]
````

````python
productos_lookup = df_productos.set_index("id_producto").to_dict("index")
clientes_ids = df_clientes["id_cliente"].tolist()
productos_ids = df_productos["id_producto"].tolist()
tiendas_ids = [t["id_tienda"] for t in tiendas]
empleados_ids = df_empleados["id_empleado"].tolist()
````

````python
ventas = []
fecha_inicio = datetime(2025, 1, 1)
fecha_fin = datetime(2026, 3, 31)
dias_rango = (fecha_fin - fecha_inicio).days

for i in range(1, N_VENTAS + 1):
    id_producto = random.choice(productos_ids)
    id_cliente = random.choice(clientes_ids)
    id_tienda = random.choice(tiendas_ids)
    id_empleado = random.choice(empleados_ids)

    producto = productos_lookup[id_producto]
    empleado = df_empleados.loc[df_empleados["id_empleado"] == id_empleado].iloc[0]

    fecha_venta = fecha_inicio + timedelta(days=random.randint(0, dias_rango))
    cantidad = random.randint(1, 8)

    precio_base = float(producto["precio_lista"])
    variacion = random.uniform(0.90, 1.05)
    precio_unitario = round(precio_base * variacion, 2)

    venta_bruta = cantidad * precio_unitario
    descuento = round(venta_bruta * random.choice([0, 0.05, 0.10, 0.15]), 2)

    ventas.append({
        "id_venta": i,
        "fecha_venta": fecha_venta.date(),
        "id_cliente": id_cliente,
        "id_producto": id_producto,
        "id_tienda": id_tienda,
        "id_empleado": id_empleado,
        "nombre_empleado": empleado["nombre_empleado"],
        "cargo": empleado["cargo"],
        "cantidad": cantidad,
        "precio_unitario": precio_unitario,
        "descuento": descuento,
        "metodo_pago": random.choice(METODOS_PAGO),
        "canal_venta": random.choice(CANALES_VENTA),
        "estado_venta": random.choice(ESTADOS_VENTA)
    })

df_ventas = pd.DataFrame(ventas)
df_ventas.head()
````

````python
print("Clientes:", df_clientes.shape)
print("Productos:", df_productos.shape)
print("Empleados:", df_empleados.shape)
print("Ventas:", df_ventas.shape)
print("Tiendas:", len(tiendas))

print("\nVentas nulas por columna:")
print(df_ventas.isnull().sum())

print("\nEstados de venta:")
print(df_ventas["estado_venta"].value_counts())

print("\nProductos referenciados inexistentes en ventas:",
      ~df_ventas["id_producto"].isin(df_productos["id_producto"]).all())

print("Clientes referenciados inexistentes en ventas:",
      ~df_ventas["id_cliente"].isin(df_clientes["id_cliente"]).all())
````

# Cargar datos en MySQL, PostgreSQL y MongoDB

````python
%pip install sqlalchemy pymysql psycopg2-binary pymongo
````

````python
from sqlalchemy import create_engine
from pymongo import MongoClient
import pandas as pd
import json
````

````python
df_clientes = pd.read_csv("../data/clientes.csv")
df_productos = pd.read_csv("../data/productos.csv")
df_ventas = pd.read_csv("../data/ventas.csv")

with open("../data/tiendas.json", "r", encoding="utf-8") as f:
    tiendas = json.load(f)
````

````python
mysql_engine = create_engine("mysql+pymysql://root:root@localhost:3307/origen_mysql")

df_clientes.to_sql(
    "clientes",
    con=mysql_engine,
    if_exists="replace",
    index=False
)

print("Clientes cargados en MySQL.")
````

````python
pd.read_sql("SELECT COUNT(*) AS total_clientes FROM clientes", mysql_engine)
````

````python
pg_engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5432/origen_pg")

df_productos.to_sql(
    "productos",
    con=pg_engine,
    if_exists="replace",
    index=False
)

print("Productos cargados en PostgreSQL.")
````

````python
pd.read_sql('SELECT COUNT(*) AS total_productos FROM productos', pg_engine)
````

````python
mongo_client = MongoClient("mongodb://localhost:27017/")
mongo_db = mongo_client["origen_mongo"]
mongo_collection = mongo_db["tiendas"]

mongo_collection.delete_many({})
mongo_collection.insert_many(tiendas)

print("Tiendas cargadas en MongoDB.")
````

````python
mongo_collection.count_documents({})
````

````python
print("Clientes MySQL:", pd.read_sql("SELECT COUNT(*) AS total FROM clientes", mysql_engine)["total"][0])
print("Productos PostgreSQL:", pd.read_sql("SELECT COUNT(*) AS total FROM productos", pg_engine)["total"][0])
print("Tiendas MongoDB:", mongo_collection.count_documents({}))
print("Ventas CSV:", len(df_ventas))
````
